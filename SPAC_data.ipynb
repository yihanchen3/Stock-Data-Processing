   - **Calculate Close-to-Close Return**: 
   
      This method calculates the percentage change in the closing price of a stock from one day to the next.
   
      $$ \text{Return}_t = \frac{\text{Close Price}_t - \text{Close Price}_{t-1}}{\text{Close Price}_{t-1}} $$

      In Python using pandas:   `df['Return'] = df['Close'].pct_change()`

   - **Map Industry Categories of Stocks to GICS Sectors**: 

      The Global Industry Classification Standard (GICS) is an industry taxonomy, consisting of 11 sectors. The stocks are grouped into 11 sectors according to their industry categories.

   - **Plot the SPACs' daily return and close price in GICs Sectors**

```python
# calculate the daily returns. Denoted as percentage, close on x - close on(x-1)/ close on (x-1) ---- so no data on the first day
df_return = (df_price.pct_change()*100).T
df_return.head(10)
```

### Third Part: Acquire Market benchmark, calculate Abnormal Return and forcasting

   - **Acquire VIX and RUSSELL 2000 data**:
   
      Extract the VIX and RUSSELL 2000 data using `yfinance` within the same time period as the SPACs data.
      Plot the VIX and SPAC returns over time.

   - **Calculate the abnormal returns and cumulative AR of SPACs**:

      Abnormal Return (AR) is the difference between the actual return of a stock and the expected return. The expected return is estimated using a market model, such as CAPM. The abnormal return is calculated as follows:

      $$ \text{AR}_t = \text{Return}_t - \hat{\alpha} - \hat{\beta} \times \text{Market Return}_t $$

      Cumulative Abnormal Return (CAR) is the sum of abnormal returns over a period of time. The cumulative abnormal return is calculated as follows:

      $$ \text{CAR}_t = \sum_{i=1}^{t} \text{AR}_i $$

      Buy-and-hold abnormal return (BHAR) is the cumulative abnormal return from the beginning of the event window to the end of the measurement period. The BHAR is calculated as follows:

      $$ \text{BHAR} = \sum_{i=1}^{T} \text{AR}_i $$

   - **Regression Analysis**:

      Regression analysis is used to analyze the relationship between SPAC returns and market benchmarks. The regression model is estimated using the Ordinary Least Squares (OLS) method. The coefficients and significance levels of the regression models are calculated using the `statsmodels` package.

   - **Forcasting SPAC Returns**:

      Autoregressive Integrated Moving Average (ARIMA) and Vector Autoregression (VAR) models are used to forecast the future returns of SPACs. The ARIMA model is a univariate time series model, while the VAR model is a multivariate time series model. 

      - ARIMA Model:
         Autocorrelation and Partial Autocorrelation Analysis are used to identify potential ARIMA models using `pt.autocorr_plot()` and `pt.plot_pacf()`.
         Find the optimal lag order of the ARIMA model using the Akaike Information Criterion (AIC).  
         The ARIMA model is estimated using the `statsmodels` package.
      - VAR Model:
         Find the optimal lag order of the VAR model using the Akaike Information Criterion (AIC). The VAR model is estimated using the `statsmodels.tsa.vector_ar.var_model` package.

      The forecasted returns for the next month are plotted using both models.

```python
# acquire the data of Russell 2000 and VIX as the market benchmark
df_russell_vix = pd.DataFrame()
russell_data = yf.download('^RUT', start=latest_completion_date, end= set_end_date)
vix_data = yf.download('^VIX', start=latest_completion_date, end= set_end_date)

df_russell_vix['RUSSELL'] = russell_data['Close'].pct_change() * 100
df_russell_vix['VIX'] = vix_data['Close']

df_russell_vix
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting Share_Price with ax1
ax1.set_xlabel('Date')
ax1.set_ylabel('Share Price', color='tab:blue')
ax1.plot(df_price_gics.mean(), color='tab:blue', label='Share Price')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Creating a second y-axis for Volume
ax2 = ax1.twinx()
ax2.set_ylabel('VIX', color='tab:orange')
ax2.plot(df_russell_vix['VIX'], color='tab:orange', alpha=0.6, label='VIX')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Adding legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')


fig.tight_layout()
plt.title('SPAC Share Price and VIX Over Time')
plt.grid(True)
plt.show()
```

```python
cumulative_length = 10

# Calculate mean return for SPAC & Merge with Russell and VIX data
df_return_russell_vix = pd.merge(df_return.mean().to_frame(name='SPAC'), df_russell_vix, on='Date').dropna()

# Calculate Abnormal Return, Cumulative AR, and Buy-and-Hold AR
df_AR = df_return_russell_vix.copy()
df_AR['Abnormal Return'] = df_AR['SPAC'] - df_AR['RUSSELL']
df_AR['Cumulative AR'] = df_AR['Abnormal Return'].rolling(window=cumulative_length).sum()
df_AR['Buy-and-Hold AR'] = (1 + df_AR['SPAC']).rolling(window=cumulative_length).apply(lambda x: x.prod()) - (1 + df_AR['RUSSELL']).rolling(window=cumulative_length).apply(lambda x: x.prod())
df_AR = df_AR.dropna()

df_AR.head(10)
```

   - **Forcasting SPAC Returns**:

      Autoregressive Integrated Moving Average (ARIMA) and Vector Autoregression (VAR) models are used to forecast the future returns of SPACs. The ARIMA model is a univariate time series model, while the VAR model is a multivariate time series model. 

      - ARIMA Model:
         Autocorrelation and Partial Autocorrelation Analysis are used to identify potential ARIMA models using `pt.autocorr_plot()` and `pt.plot_pacf()`.
         Find the optimal lag order of the ARIMA model using the Akaike Information Criterion (AIC).  
         The ARIMA model is estimated using the `statsmodels` package.
      - VAR Model:
         Find the optimal lag order of the VAR model using the Akaike Information Criterion (AIC). The VAR model is estimated using the `statsmodels.tsa.vector_ar.var_model` package.

      The forecasted returns for the next month are plotted using both models.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

plot_acf(df_AR['SPAC'], lags=40, ax=ax1)
ax1.set_title('ACF for SPAC Returns')
plot_pacf(df_AR['SPAC'], lags=40, ax=ax2)
ax2.set_title('PACF for SPAC Returns')

plt.tight_layout()
plt.show()
```

```python
p_q_order = md.order_ARIMA(df_AR['SPAC'])
forecasted_returns, conf_int = md.forecast_ARIMA(df_AR['SPAC'], order=p_q_order, forecast_steps=30)
print('Model Coefficients:', conf_int)

# Plotting 
fig = plt.figure(figsize=(12, 4))
plt.plot(forecasted_returns, label = 'SPAC Return')
plt.title("Forecasted returns for the next month (30 days) using ARIMA" + str(p_q_order))
plt.xlabel("Days")
plt.ylabel("SPAC Return")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```
